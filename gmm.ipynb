```python
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
X, y_true = make_blobs(n_samples=300, centers=3, n_features=2, 
                       cluster_std=0.6, random_state=42)

# Fit Gaussian Mixture Model
n_components = 3
gmm = GaussianMixture(n_components=n_components, random_state=42)
gmm.fit(X)

# Get cluster labels and probabilities
labels = gmm.predict(X)
proba = gmm.predict_proba(X)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Clustered data
ax1 = axes[0]
scatter = ax1.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6, edgecolors='k')
ax1.scatter(gmm.means_[:, 0], gmm.means_[:, 1], c='red', marker='X', s=200, edgecolors='k', linewidths=2)
ax1.set_title('GMM Clustering Results')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
plt.colorbar(scatter, ax=ax1, label='Cluster')

# Plot 2: Cluster probabilities for first 100 samples
ax2 = axes[1]
im = ax2.imshow(proba[:100], cmap='viridis', aspect='auto')
ax2.set_title('Cluster Membership Probabilities')
ax2.set_xlabel('Cluster')
ax2.set_ylabel('Sample')
plt.colorbar(im, ax=ax2, label='Probability')

plt.tight_layout()
plt.show()

# Print model information
print(f"AIC: {gmm.aic(X):.2f}")
print(f"BIC: {gmm.bic(X):.2f}")
print(f"Log-likelihood: {gmm.score(X):.2f}")
```
